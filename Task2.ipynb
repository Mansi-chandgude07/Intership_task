### Removing Unnecessary data from dataset

### Finding missing values and handling them

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median()).astype(int)
data['Utilization_Ratio'].fillna(data['Utilization_Ratio'].median(), inplace=True)
data['Credit_Score'].fillna(data['Credit_Score'].median(), inplace=True)
data['Annual_Income'].fillna(data['Annual_Income'].median(), inplace=True)
data['Employment_Type'].fillna(
    data['Employment_Type'].mode()[0],
    inplace=True
)
#numeric are handle by median because our dataset it not normally distributeded
```

## Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

The distribution plot shows the number of customers who defaulted versus those who did not. The dataset appears slightly imbalanced, with non-defaulters being more frequent. This indicates that default cases form a minority class, which may require special handling during model training.

```python
sns.countplot(x='Default_Status', data=data)
plt.title("Default Status Distribution")
plt.xlabel("Default (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

The histogram shows that Annual Income is right-skewed, meaning most applicants fall in the lower to mid-income range, with a few high-income outliers. This suggests income inequality among applicants and indicates that income alone may not be sufficient to predict default risk.

```python
plt.hist(data['Annual_Income'], bins=30)
plt.title("Distribution of Annual Income")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

The boxplot compares loan amounts between defaulters and non-defaulters. It appears that defaulters tend to have slightly higher loan amounts and more extreme outliers. This suggests that higher loan amounts may increase the probability of default

```python
sns.boxplot(x='Default_Status', y='Loan_or_Credit_Amount', data=data)
plt.title("Loan Amount vs Default Status")
plt.show()
```

The correlation heatmap shows relationships between numerical features. Credit Score appears negatively correlated with Default Status, while Utilization Ratio and Past Default Count show positive correlation. This indicates that lower credit scores and higher utilization increase default risk.

```python
plt.figure(figsize=(10,8))

numeric_df = data.select_dtypes(include=['int64','float64'])
corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

The pairplot visualizes relationships between Age, Annual Income, Credit Score, and Default Status. Some separation between default and non-default customers is visible, especially in Credit Score distribution. However, other features show overlap, suggesting multiple variables are needed for accurate prediction.

```python

sns.pairplot(data[['Age','Annual_Income','Credit_Score','Default_Status']], 
             hue='Default_Status')
plt.show()
```

The swarm plot clearly shows that defaulters generally have lower credit scores compared to non-defaulters. This confirms that credit score is a strong predictor of default behavior


```python
sns.swarmplot(x='Default_Status', y='Credit_Score', data=data)
plt.title("Credit Score vs Default")
plt.show()
```
