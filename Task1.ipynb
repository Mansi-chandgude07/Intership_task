### Removing Unnecessary data from dataset

```python

data = data.drop(['Customer_ID', 'Application_Date'], axis=1)
```

### Finding missing values and handling them

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median()).astype(int)
data['Utilization_Ratio'].fillna(data['Utilization_Ratio'].median(), inplace=True)
data['Credit_Score'].fillna(data['Credit_Score'].median(), inplace=True)
data['Annual_Income'].fillna(data['Annual_Income'].median(), inplace=True)
data['Employment_Type'].fillna(
    data['Employment_Type'].mode()[0],
    inplace=True
)
#numeric are handle by median because our dataset it not normally distributeded
```
